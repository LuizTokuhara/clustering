```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler
from sklearn.metrics import silhouette_score, silhouette_samples, pairwise_distances_argmin_min, make_scorer
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
```

```python
# Carregar dataset
df = pd.read_csv('../dataset/updated_pollution_dataset.csv', sep=',', decimal='.')

# Verificando dados
df.head(10)
```

<h1>Análise dos dados

```python
unique_values = df['Air Quality'].unique()
print(f"Valores únicos na coluna 'Air Quality': {unique_values}")

value_counts = df['Air Quality'].value_counts()
print(f"Contagem de cada valor na coluna 'Air Quality':\n{value_counts}")
```

```python
# Estatísiticas do dataset
df.describe()
```

```python
# Plotagem do dataset
df.hist(figsize=(15, 10), bins=15)
plt.show()
```

```python
df['Air Quality'] = df['Air Quality'].replace({'Moderate': 0, 'Good': 1, 'Hazardous': 2, 'Poor': 3})
corr = df.corr()
corr_target = corr['Air Quality']
corr_target.sort_values(ascending = False)
corr_target
```

```python
features = df.drop(['Air Quality'], axis=1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
scaled_data
```

<h1>Análise da quantidade de clusters via dendograma

```python
linked = linkage(scaled_data, method='ward')  # 'ward' minimiza variância

# Plot dendograma
plt.figure(figsize=(15, 10))
dendrogram(
    linked, 
    orientation='top', 
    distance_sort='descending', 
    show_leaf_counts=True, 
    no_labels=True  # Remove os labels inferiores
)
plt.title('Dendrograma')
plt.xlabel('Pontos')
plt.ylabel('Distância Euclidiana')
plt.hlines(y=65,xmin=0,xmax=50000,lw=2,linestyles='--',colors='black')
plt.text(14000, 68, 'Corte ideal indica 4 clusters', ha='left', va='center')
plt.show()
```

<h1>Análise da quantidade de clusters com Elbow

```python
# Encontrar o melhor número de clusters
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(scaled_data)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 10), inertia, marker="o")
plt.xlabel("Número de clusters")
plt.ylabel("Inertia")
plt.title("Elbow")
plt.show()
```

<h1>K-Means

```python
# Aplicando o melhor resultado baseado no dendograma e elbow
k_cluster = 4
kmeans = KMeans(n_clusters=k_cluster,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(scaled_data)
silhouette_avg = silhouette_score(scaled_data, y_kmeans)
print(f"Silhouette Score for k:{k_cluster} = {silhouette_avg}")
```

```python
# Calcula score da silhueta
silhouette_avg = silhouette_score(scaled_data, kmeans.labels_)
print(f'Silhouette Score for K={k_cluster}: {silhouette_avg:.4f}')

sample_silhouette_values = silhouette_samples(scaled_data, kmeans.labels_)

plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(k_cluster):
    ith_cluster_silhouette_values = sample_silhouette_values[kmeans.labels_ == i]
    
    ith_cluster_silhouette_values.sort()
    
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    
    color = sns.color_palette("deep")[i]
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)
    
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

plt.axvline(x=silhouette_avg, color="red", linestyle="--")
plt.title("Gráfico de silhueta para agrupamento de K-means")
plt.xlabel("Valores do coeficiente de silhueta")
plt.ylabel("Label do cluster")
plt.show()
```

<h1>Verificação do agrupamento dos clusters

```python
# Redução de dimensionalidade somente para visualização
pca = PCA(n_components=2, random_state=42)
pca_components = pca.fit_transform(scaled_data)

# Dataframe com os componentes PCA
pca_df = pd.DataFrame(data=pca_components, columns=['PCA1', 'PCA2'])

# Transformar os centros dos clusters para o espaço reduzido
centroids_pca = pca.transform(kmeans.cluster_centers_)

# Criar um DataFrame combinando os dados reduzidos e os clusters
visual_df = pd.concat([pca_df, pd.DataFrame({'Cluster': kmeans.labels_})], axis=1)

# Visualizar os dados
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=visual_df, palette='viridis', s=100, alpha=0.7)
plt.title('K-Means Clusters com PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', s=200, alpha=0.7, marker='X', label='Centroids')
plt.show()
```

<h1>DBScan

```python
# Calcula as distâncias dos k-vizinhos mais próximos
nbrs = NearestNeighbors(n_neighbors=5).fit(scaled_data)
distances, indices = nbrs.kneighbors(scaled_data)

# Ordena as distâncias para o gráfico
distances = np.sort(distances[:, -1])
plt.plot(distances)
plt.xlabel("Pontos ordenados")
plt.ylabel("Distância do 5º vizinho mais próximo")
plt.title("Gráfico de distância k-dist")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# dbscan = DBSCAN(eps=1.42, min_samples=3).fit(scaled_data)  # melhor resultado na busca manual
# dbscan = DBSCAN(eps=0.4, min_samples=4).fit(scaled_data)   # melhor resultado encontrado pelo GridSearch - score negativo (ruim)
dbscan = DBSCAN(eps=2, min_samples=5).fit(scaled_data)       # resultado encontrado no vizinho mais próximos
core_samples_mask = np.zeros_like(dbscan.labels_, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
labels = dbscan.labels_

silhouette_avg = silhouette_score(scaled_data, labels)
print(f"DBScan Silhouette Score: {silhouette_avg}")

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

# Visualização 2D
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, col in zip(unique_labels, colors):
    if label == -1:
        # Cor preta para ruídos
        col = [0, 0, 0, 1]

    class_member_mask = (labels == label)
    
    # Plotando os clusters
    xy = scaled_data[class_member_mask]
    plt.plot(xy[:, 4], xy[:, 6], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title(f'DBScan clusters: {n_clusters_}')
plt.xlabel('NO2')
plt.ylabel('CO')
plt.grid(True)
plt.show()
```
